## In terms of author number, after checking the original paper PDFs, I found that ieee is always correct, except for this one:

- 10.1109/TVCG.2008.157, where IEEE website lists a first author who does not appear in the PDF. 

`ieee_df` misses info for `10.1109/VIS.1999.10000`, this is because the url directs me to a site other than ieee.org. The info on OPENALEX for this paper is correct. 

```python
def update_ieee_orig(DF): # df here is iee_orig
    """update ieee_org

    ieee_org is wrong in '10.1109/TVCG.2008.157' as it contains an additional author that shouldn't be there;
    also, ieee_org lacks author info for '10.1109/VIS.1999.10000'.

    What this function does is to delete the additional author in '10.1109/TVCG.2008.157' and update info in 
    that paper. Then, I added author data manually for '10.1109/VIS.1999.10000'.

    """
    DF = DF.drop(DF[DF.DOI == '10.1109/VIS.1999.10000'].index)
    row_to_drop = DF.index[DF.DOI == '10.1109/TVCG.2008.157'].tolist()[0]
    df_dropped = DF.drop([row_to_drop])
    df_dropped.loc[df_dropped.DOI == '10.1109/TVCG.2008.157', 'Number of Authors'] -= 1
    df_dropped.loc[df_dropped.DOI == '10.1109/TVCG.2008.157', 'Author Position'] -= 1.0
    df = df_dropped
    FILL_DATA = [
    {
        'Year': 1999,
        'DOI': '10.1109/VIS.1999.10000',
        'Title': 'Progressive Compression of Arbitrary Triangular Meshes',
        'Number of Authors': 3,
        'Author Position': 1,
        'Author Name': 'Daniel Cohen-Or',
        'Author ID': np.NaN,
        'Author Affiliation': 'Tel Aviv University',
        'One Affiliation': True,
    },
    {
        'Year': 1999,
        'DOI': '10.1109/VIS.1999.10000',
        'Title': 'Progressive Compression of Arbitrary Triangular Meshes',
        'Number of Authors': 3,
        'Author Position': 2,
        'Author Name': 'David Levin',
        'Author ID': np.NaN,
        'Author Affiliation': 'Tel Aviv University',
        'One Affiliation': True,
    },
    {
        'Year': 1999,
        'DOI': '10.1109/VIS.1999.10000',
        'Title': 'Progressive Compression of Arbitrary Triangular Meshes',
        'Number of Authors': 3,
        'Author Position': 3,
        'Author Name': 'Offir Remez',
        'Author ID': np.NaN,
        'Author Affiliation': 'Tel Aviv University',
        'One Affiliation': True,
    }
    ]
    fill_data_df = pd.DataFrame(FILL_DATA)
    df = df.append(fill_data_df, ignore_index = True)
    return df
```

## Comparing with VISPD/DPLD

```python
vispd = pd.read_csv('../../data/raw/vispubdata.csv')
```

```python
vispd = vispd[vispd.DOI.isin(dois)].reset_index(drop=True)
vispd[vispd.AuthorNames.isnull()]
vispd.head()
```

```python
vispd['AuthorNumber'] = [len(re.findall(r'\;', name)) + 1 for name in vispd['AuthorNames-Deduped'].tolist()]
```

```python
vispd_affili_complete = vispd[vispd['AuthorAffiliation'].notnull()].reset_index(drop=True)
```

```python
vispd_affili_complete['AffiliationNumber'] = [
    len(re.findall(r'\;', affiliations)) + 1 for affiliations in vispd_affili_complete[
        'AuthorAffiliation'].tolist()]
```

## Compare author number in vispd and that in merged author df

```python
vispd_author_num = vispd.AuthorNumber.tolist()
```

```python
# ieee has 2021 data but vispd doesn't
ieee = ieee[ieee.DOI.isin(vispd.DOI)].reset_index(drop=True)
```

```python
ieee.shape
```

```python
ieee['PaperIndex'] = [vispd.DOI.tolist().index(doi) for doi in ieee.DOI.tolist()]
```

```python
ieee.head()
```

```python
ieee_author_num = [
    group['Number of Authors'].tolist()[0] for paper_index, group in ieee.groupby('PaperIndex')
]
```

```python
ieee_author_num == vispd_author_num
```

```python
diff_idx_dict_list = []
for idx in range(0, len(vispd_author_num)):
    if ieee_author_num[idx] != vispd_author_num[idx]:
        diff_idx_dict = {
            'Index': idx,
            'Year': vispd.Year[idx],
            'DOI': vispd.DOI[idx],
            'Title': vispd.Title[idx],
            'Author Number in vispd': vispd_author_num[idx],
            'Author Number in ieee': ieee_author_num[idx]
        }
        diff_idx_dict_list.append(diff_idx_dict)
```
