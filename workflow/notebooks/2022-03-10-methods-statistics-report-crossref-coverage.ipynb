In this notebook, i want to check: 
  1. How many papers in `paeprs_to_study.txt` are accessible on CrossRef via doi search
  2. How many have author affiliation info.
  3. How many have `subject` info. 

```python
import requests
import pandas as pd
import json
import time
```

```python
dois = pd.read_csv('../../data/processed/vispd_plus_good_papers.txt', header=None)[0].tolist()
```

```python
prefix_str = 'https://api.crossref.org/v1/works/'
```

```python
retry_code = [ 500, 502, 503, 504, 429]
```

```python
def get_response(doi):
    response = requests.get(prefix_str + doi)
    return response
```

```python
crossref_dict_list = []
```

```python
for doi in dois:
    doi_index = dois.index(doi) + 1
    r = get_response(doi)
    while r.status_code in retry_code:
        print(f'doi query for {doi_index} : {doi} has error, status code is {r.status_code}, retrying...')
        time.sleep(3)
        r = get_response(doi)
    if r.status_code == 200:
        j = r.json()['message']
        reference_count = j['references-count']
        first_author_aff = j['author'][0]['affiliation']
        crossref_dict = {
            'DOI': doi,
            'Status Code': r.status_code,
            'Reference Count': reference_count,
            'First Author Affiliation': first_author_aff,
        }
    else:
        crossref_dict = {
            'DOI': doi,
            'Status Code': r.status_code,
        }
    crossref_dict_list.append(crossref_dict)
    print(f'{doi_index} is done')
    time.sleep(0.5)
```

```python
df = pd.DataFrame(crossref_dict_list)
df.to_csv('../../data/interim/methods_reporting/crossref.csv', index = False)
```
