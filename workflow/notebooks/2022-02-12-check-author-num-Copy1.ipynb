# Write the update to function

```python
def update_df_ieee(df, fill_data): # df here is df_ieee
    row_to_drop = df.index[df.DOI == '10.1109/TVCG.2008.157'].tolist()[0]
    df_dropped = df.drop([row_to_drop])
    df_dropped.loc[df_dropped.DOI == '10.1109/TVCG.2008.157', 'Number of Authors'] -= 1
    df_dropped.loc[df_dropped.DOI == '10.1109/TVCG.2008.157', 'Author Position'] -= 1.0
    row_to_drop = df.index[df.DOI == '10.1109/VIS.1999.10000'].tolist()[0]
    df_dropped = df_dropped.drop([row_to_drop])
    df = df_dropped
    fill_data_df = pd.DataFrame(fill_data)
    df = df.append(fill_data_df, ignore_index = True)
    return df
```

```python
df_ieee = pd.read_csv('../../data/interim/ieee_author_df.csv')
```

```python
df_ieee_updated = update_df_ieee(df_ieee, fill_data)
```
