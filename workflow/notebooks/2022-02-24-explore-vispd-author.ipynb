## Aim

In this notebook, I want to see:

- Whether I can use the author affiliation data in VISPUBDATA. To do that, I need to first check whether the number of author affiliations match that of author names. 
- Whether the number of authors in VISPUBDATA is equal to that of merged author df . 

```python
import pandas as pd
import csv
import re
```

```python
def read_txt(INPUT):
	"""read txt files and return a list
	"""
	raw = open(INPUT, "r")
	reader = csv.reader(raw)
	allRows = [row for row in reader]
	data = [i[0] for i in allRows]
	return data
```

```python
vispd = pd.read_csv('../../data/raw/vispubdata.csv')
DOIS = read_txt('../../data/processed/papers_to_study.txt')
```

```python
vispd = vispd[vispd.DOI.isin(DOIS)].reset_index(drop=True)
vispd[vispd.AuthorNames.isnull()]
vispd
```

```python
vispd['AuthorNumber'] = [len(re.findall(r'\;', name)) + 1 for name in vispd['AuthorNames-Deduped'].tolist()]
```

```python
vispd_affili_complete = vispd[vispd['AuthorAffiliation'].notnull()].reset_index(drop=True)
```

```python
vispd_affili_complete['AffiliationNumber'] = [
    len(re.findall(r'\;', affiliations)) + 1 for affiliations in vispd_affili_complete['AuthorAffiliation'].tolist()]
```

## In rows where VISPD has affiliations, check whether the number of affiliation is equal to the number of aturhos.

Turns out, in 1/4 of the rows, they are not equal. 

```python
vispd_affili_complete[vispd_affili_complete.AuthorNumber != vispd_affili_complete.AffiliationNumber]
```

## Compare author number in vispd and that in merged author df

```python
vispd_author_num = vispd.AuthorNumber.tolist()
```

```python
merged_df = pd.read_csv('../../data/processed/merged_author_df.csv')
```

```python
merged_df = merged_df[merged_df.DOI.isin(vispd.DOI)].reset_index(drop=True)
```

```python
merged_df['PaperIndex'] = [vispd.DOI.tolist().index(doi) for doi in merged_df.DOI.tolist()]
```

```python
merged_author_num = [
    group['IEEE Number of Authors'].tolist()[0] for doi, group in merged_df.groupby('PaperIndex')
]
```

```python
merged_author_num == vispd_author_num
```

```python
diff_idx_dict_list = []
for idx in range(0, len(vispd_author_num)):
    if merged_author_num[idx] != vispd_author_num[idx]:
        diff_idx_dict = {
            'Index': idx,
            'Year': vispd.Year[idx],
            'DOI': vispd.DOI[idx],
            'Title': vispd.Title[idx],
            'Author Number in vispd': vispd_author_num[idx],
            'Author Number in merged': merged_author_num[idx]
        }
        diff_idx_dict_list.append(diff_idx_dict)
```

```python
pd.DataFrame(diff_idx_dict_list)
```
