## Aim

The aim of this notebook is to show the top 100 venues citing VIS papers. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('openalex_citation_paper_df.csv')
```

```python
len(df)
```

```python
venues = df['OpenAlex Venue Name']
```

```python
venues[1]
```

```python
l = Counter(venues).most_common(100)
l[1][1]/len(df)
```

```python
l = Counter(venues).most_common(100)
top_df = pd.DataFrame(l, columns = ["Journal/Conference", "Counts"])
top_df.index = np.arange(1, len(top_df) + 1)
```

```python
counts = top_df['Counts'].tolist()
# improved_counts = [str(c) + " (" + "{0:.2%}".format(c/len(df))+")" for c in counts]
perc = ["{0:.2%}".format(c/len(df)) for c in counts]
top_df['percentage'] = perc
```

```python
top_df.style.bar(subset=['Counts'], color='orange')
```
